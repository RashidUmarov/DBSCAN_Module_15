```python
import pandas as pd
```

```python
df = pd.read_csv('gt_2015.csv')
df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
df_scaled.head()
```

```python
from sklearn.cluster import DBSCAN

model = DBSCAN(eps=0.5, min_samples=5)
model.fit(df_scaled)
```

```python
df['labels'] = model.labels_
df.head()
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(df_scaled, df['labels'])
```

```python
df['labels'].value_counts()
```

**Вывод**: Слишком много кластеров (122) и выбросов  (кластер с меткой -1)

### Изменим гиперпараметры

```python
model = DBSCAN(eps=1, min_samples=5)
model.fit(df_scaled)
```

```python
df['labels'] = model.labels_
df.head()
```

```python
silhouette_score(df_scaled, df['labels'])
```

```python
df['labels'].value_counts()
```

**Вывод**: почти все точки попали в один кластер, количество кластеров резко снизилос. Количество выбросов тоже

### Уменьшим размер окрестности

```python
model = DBSCAN(eps=0.7, min_samples=5)
model.fit(df_scaled)
```

```python
df['labels'] = model.labels_
df.head()
```

```python
silhouette_score(df_scaled, df['labels'])
```

```python
df['labels'].value_counts()
```
